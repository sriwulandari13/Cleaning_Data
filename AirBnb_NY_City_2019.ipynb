```python
import pandas as pd
```

```python
airbnb = pd.read_csv('AB_NYC_2019.csv')
airbnb.info()
```

```python
#menghapus data duplicate
airbnb = airbnb.drop_duplicates()
```

```python
#ubah type data kolom last_review
airbnb["last_review"]
```

```python
airbnb["last_review"] = pd.to_datetime(airbnb["last_review"])
```

```python
airbnb.info()
```

```python
airbnb["last_review"]
```

```python
#cek missing value di kolom host_name
airbnb[airbnb["host_name"].isna()]
```

```python
#drop data dari host_name yang missing
airbnb = airbnb[~airbnb["host_name"].isna()]
airbnb.info()
```

```python
#kolom name masih ada missing value
airbnb[airbnb["name"].isna()]
```

```python
airbnb = airbnb[~airbnb["name"].isna()]
airbnb.info()
```

```python
#cek data outliers di kolom price
airbnb.boxplot(column="price")
```

```python
#cek data outliers kolom price menggunakan statistik
airbnb["price"].describe()
```

Dapat dilihat dari statistik diatas, jika nilai q1 q2 dan q3 tidak seimbang thd nilai min dan max maka dapat disimpulkan bahwa data outliers, next step ke pembersihan data outliers

```python
q1 = airbnb['price'].quantile(0.25)
q3 = airbnb['price'].quantile(0.75)
upper_limit = q3 + 1.5 * (q3-q1)
lower_limit = q1 - 1.5 * (q3-q1)
```

```python
airbnb = airbnb[airbnb['price'] > lower_limit]
airbnb = airbnb[airbnb['price'] < upper_limit]
airbnb['price'].describe()
```

```python
#join kolom latitude & longitude
#karena typedata kolom latitude & longitude berupa float/angka, maka ubah type data ke str dulu
airbnb['latitude'] = airbnb['latitude'].astype('str')
airbnb['longitude'] = airbnb['longitude'].astype('str')
airbnb.info()
```

```python
#buat kolom baru untuk penggabungan latitude & longitude
airbnb['Lat, Long'] = airbnb['latitude']+","+airbnb['longitude']
airbnb.head(3)
```

```python
#ubah type data id & host_id ke string
airbnb['id'] = airbnb['id'].astype('str')
airbnb['host_id'] = airbnb['host_id'].astype('str')
```
